```python
import os
import json

DATA_DIR = os.path.join('data', 'boxofficemojo')

target_file_names = os.listdir(DATA_DIR)
movie = list()
for files in target_file_names:
    target_file_path = os.path.join(DATA_DIR, files)
    with open(target_file_path,'r') as target_file:
        movie.append(json.load(target_file))
        
len(movie)
```

```python
from collections import Counter

movie_years = [d['year'] for d in movie]
movie_counter = Counter(movie_years)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

movie_years_filter = filter(None, movie_years)
plt.hist(movie_years_filter)
plt.title('Your title goes here');
```

```python
df = movies_df[['opening_weekend_take','domestic_gross']]
```

```python
df = df.dropna()
```

```python
len(df)
```

```python
import statsmodels.api as sm
Y = df.domestic_gross
X = sm.add_constant(df.opening_weekend_take)

linmodel = sm.OLS(Y,X).fit()
```

```python
linmodel.summary()
```

```python
predicted_gross = linmodel.predict(X)
plt.scatter(df.opening_weekend_take, df.domestic_gross, color='gray')
plt.plot(df.opening_weekend_take, predicted_gross)
plt.title("Opening Weekend Based Model for Domestic Gross")
plt.xlabel("Opening Weekend Take ($100M)")
plt.ylabel("Domestic Gross ($100M)")
```

```python
plt.scatter(df.domestic_gross,predicted_gross,color='gray')
plt.plot(df.domestic_gross,df.domestic_gross)
```

```python
residuals = predicted_gross - df.domestic_gross
plt.scatter(predicted_gross,residuals,color='gray')
plt.plot([min(predicted_gross),max(predicted_gross)],[0,0])
plt.title("Predicted Gross vs Residuals")
plt.xlabel("Predicted Domestic Gross")
plt.ylabel("Residuals")
```

```python
Y_log = np.log(df.domestic_gross)
X_log = sm.add_constant(np.log(df.opening_weekend_take))

linmodel_log = sm.OLS(Y_log,X_log).fit()
linmodel_log.summary()

predicted_gross_log = linmodel_log.predict(X_log)
plt.scatter(np.log(df.opening_weekend_take), np.log(df.domestic_gross), color='gray')
plt.plot(np.log(df.opening_weekend_take), predicted_gross_log)
plt.title("Opening Weekend Based Model for Domestic Gross")
plt.xlabel("Opening Weekend Take")
plt.ylabel("Domestic Gross")
```
